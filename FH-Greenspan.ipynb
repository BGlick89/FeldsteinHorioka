# Feldstein-Horioka Analysis Tools

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from linearmodels.panel import PanelOLS
import seaborn as sns
import numpy as np
from scipy import stats
```

```python
# Read the Excel file into a DataFrame
df = pd.read_stata('green.dta')
```

```python

# Example data:
data = df[['year', 'sigma']]

# Sort the data by 'year'
data = data.sort_values(by='year')

# Extract 'year' and 'sigma' columns
years = data['year']
sigmas = data['sigma']  # Divide all 'sigma' values by 100

# Create a larger figure
plt.figure(figsize=(10, 6))

# Create a bar chart without data labels
plt.bar(years, sigmas, color='skyblue', edgecolor='black', width=0.6)

# Add grid lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Customize axis labels
plt.xlabel('Year', fontsize=12)
plt.ylabel('Dispersion (Billions, $) ', fontsize=12)  # Updated y-axis label

# Add a title
plt.title('Dispersion of World Current Account', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the bar chart
plt.show()
```
